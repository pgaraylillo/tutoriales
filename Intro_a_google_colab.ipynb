## 1. ¿Qué es Google Colab?

**Google Colab** (oficialmente *Colaboratory*) es un servicio gratuito de Google que te permite escribir y ejecutar código Python **directamente en el navegador**, sin instalar absolutamente nada en tu computador.

Lo que hace especial a Colab no es Python — es que te entrega un **computador virtual** con todo preinstalado (incluyendo GPU si la pides), listo para usar en 5 segundos.


## 2. ¿Por qué Colab es importante si quieres aprender IA?

Esta es la pregunta clave. Si vas a invertir tiempo en aprender Colab, ¿por qué vale la pena?

### 🎯 Razón 1: Es el estándar para enseñar y compartir IA

Cuando OpenAI, Google, Hugging Face o cualquier paper de IA publica un ejemplo, lo más probable es que venga **como un notebook de Colab**. Si no sabes abrirlos, te quedas afuera de la mayoría del material gratuito disponible.

### 💻 Razón 2: GPU gratis (o por muy poca plata)

Entrenar o usar modelos de IA requiere GPU. Una GPU decente cuesta sobre USD 1.500 y tu notebook clínico no la tiene. Colab te presta una **gratis** por hasta unas horas al día. Para aprender, sobra. Para más uso intensivo, Colab Pro cuesta ~10 USD/mes y te da GPUs mejores.

### 🚫 Razón 3: Cero instalación, cero dramas

Configurar Python, CUDA, drivers de GPU, librerías que pelean entre versiones... es la razón #1 por la que la gente se rinde aprendiendo IA. Colab elimina todo ese sufrimiento: abres el navegador y ya está listo.

### 🔁 Razón 4: Reproducibilidad

Un notebook de Colab que te funciona a ti, **le funciona igual a tu colega** que lo abre desde otro computador. Esto es enorme para enseñar, colaborar en investigación, o compartir un experimento clínico-IA con tu equipo.

### 📤 Razón 5: Compartir es trivial

Igual que un Google Doc: le mandas el link a alguien y lo puede ver, copiar y ejecutar. Sin "descarga estos 12 archivos y configura este entorno".

### 🩺 Razón 6 (la que importa para clínicos)

Si quieres explorar IA en medicina sin volverte ingeniero, Colab te permite **probar herramientas reales en problemas reales** — transcribir reuniones, analizar bases de datos clínicas anonimizadas, prototipar un asistente — sin tener que pedirle nada a IT.

## 3. ¿Qué es un *notebook* y en qué se diferencia de un script?

Un **script** tradicional es un archivo de código que corre de principio a fin, como un protocolo: lo ejecutas y hace todo de una.

Un **notebook** es distinto: está dividido en **celdas** que se ejecutan **una a la vez**, y entre celdas puedes escribir texto explicativo, ver gráficos intermedios, modificar valores y volver a correr.

> 💡 **Analogía clínica:** un script es como una orden médica con todos los pasos escritos para enfermería. Un notebook es como pasar visita: examinas, observas, tomas decisiones, anotas, sigues. Cada celda es un "paso de la visita".

Hay dos tipos de celdas:

- **Celdas de texto (Markdown):** como esta. Se usan para explicar, documentar, poner títulos.
- **Celdas de código:** ejecutan Python.

Para ejecutar una celda: la seleccionas y aprietas **`Shift + Enter`** (o el botón ▶️ a la izquierda).

## 5. El "runtime" (motor que ejecuta tu código)

Cuando abres un notebook de Colab, Google te asigna una **máquina virtual temporal** llamada **runtime** (o *kernel*). Esa máquina:

- Tiene Python y cientos de librerías ya instaladas
- Tiene su propia memoria, disco y (opcionalmente) GPU
- **Se desconecta sola** después de un rato sin uso (~90 minutos de inactividad)
- **Se borra completamente** al cerrarse — todo lo que hayas instalado o descargado en ella desaparece

> ⚠️ **Esto es crítico de entender:** el almacenamiento del runtime es **efímero**. Si descargas un archivo o entrenas un modelo y no lo guardas en Drive, lo pierdes cuando se desconecte.

Por eso casi todos los notebooks empiezan con dos cosas: **instalar las librerías que necesitan** y **conectar Google Drive**.

## 6. Instalar paquetes: el signo `!`

En una celda normal escribes Python. Pero si quieres correr un comando del **sistema operativo** (instalar algo, listar archivos, etc.), antepones un signo de exclamación `!`.

Para instalar una librería de Python se usa `pip install`:


```python
# El '!' le dice a Colab: 'esto no es Python, es un comando del sistema'
!pip install -q rich
```

La bandera `-q` significa *quiet* (silencioso) — no llena la pantalla con logs de instalación.

Ahora podemos usar la librería que acabamos de instalar:


```python
from rich import print as rprint
from rich.panel import Panel

rprint(Panel.fit(
    "[bold green]✅ Acabas de instalar y usar una librería externa[/bold green]\n"
    "[dim]Esto es lo mismo que harías para instalar herramientas de IA[/dim]",
    title="🎉",
))
```

> 💡 **Analogía clínica:** `pip install` es como pedir un fármaco a farmacia. La primera vez tarda en llegar (se instala), pero una vez que está, lo usas todas las veces que quieras dentro de esta sesión. Si la sesión se cierra, hay que volver a pedirlo.

## 8. Activar GPU (gratis)

Por defecto, Colab te da un runtime **sin GPU** (CPU solamente). Para casi cualquier modelo de IA moderno necesitas GPU. Activarla es así:

1. Menú superior → **`Entorno de ejecución`** (o *Runtime* en inglés)
2. **`Cambiar tipo de entorno de ejecución`**
3. En "Acelerador por hardware" elige **`T4 GPU`** (la opción gratis)
4. Guardar

⚠️ **Ojo:** cambiar el tipo de runtime **reinicia la sesión**. Pierdes las variables y librerías instaladas. Por eso conviene hacerlo *antes* de empezar a trabajar.

Una vez activada, esta celda debería confirmártelo:


```python
import torch

if torch.cuda.is_available():
    print(f"✅ GPU activa: {torch.cuda.get_device_name(0)}")
    print(f"   Memoria: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")
else:
    print("⚠️ Sin GPU. Anda al menú: Entorno de ejecución → Cambiar tipo → T4 GPU")
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

## 9. Conectar Google Drive (tu disco persistente)

Como vimos, el runtime es efímero. La solución estándar es **montar tu Google Drive** dentro del runtime, así cualquier archivo que guardes en Drive sobrevive entre sesiones.

La celda de abajo te abrirá una ventana para autorizar el acceso. **Solo te pedirá tu propia cuenta de Google** — Colab nunca ve tus archivos directamente, solo te permite leerlos/escribirlos desde el runtime.


```python
from google.colab import drive
drive.mount('/content/drive')
```

Después de aceptar, tu Drive aparece como una carpeta en `/content/drive/MyDrive/`. La puedes usar como cualquier carpeta:


> 🔒 **Sobre privacidad:** si trabajas con datos clínicos, **nunca** los subas a tu Drive personal. Usa una cuenta institucional o, mejor, mantén los datos en infraestructura de tu institución. Colab corre en servidores de Google y los datos pasan por ahí.

## 11. Limitaciones que tienes que conocer

Colab es gratis y eso tiene un costo:

- ⏱️ **Sesiones de máximo ~12 horas** (luego se cierra y hay que reconectarse).
- 💤 **Desconexión por inactividad** (~90 min sin tocarlo).
- 🎮 **Cuota diaria de GPU.** Si la usas mucho un día, te bloquea por unas horas.
- 🐢 **GPU compartida.** En horas peak puede que no haya T4 disponible.
- 💾 **Almacenamiento efímero.** Todo lo que no esté en Drive se pierde.
- 🔓 **No es para datos sensibles.** Tus datos pasan por servidores de Google.

Para uso serio (entrenar modelos por días, datos sensibles, producción) hay alternativas: **Kaggle Notebooks** (similar, también gratis), **Colab Pro/Pro+**, o servicios pagos como **AWS SageMaker**, **Google Vertex AI**, **Azure ML**.

Pero para **aprender, prototipar y demostrar** — Colab es imbatible.
